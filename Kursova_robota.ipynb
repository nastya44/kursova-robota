```python
import numpy as np
import sympy as sp
from math import sqrt
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
def func(fx,x1,y1):
  fx = fx.subs(x,x1)
  fx = fx.subs(y,y1)
  return fx
```

Метод внутрішної точки.

```python
start_x = 1.1
start_y = 0.1
old_x = start_x
old_y = start_y
a = 10
b = 0.2
el = 0.01
it_amount = 0
r=1
x = sp.Symbol('x',real = True)
y = sp.Symbol('y',real = True)
fx = (100*((y - x**2)**2)) + ((1-x)**2) + r*sp.ln(y-1.5)

while(True):
  while(True):
    if it_amount > 100:
      #print('Amount of it: ',it_amount)
      break
    it_amount = it_amount + 1
    px = np.matrix([[start_x],[start_y]],dtype='float')
    gx = fx.diff(x)
    gy = fx.diff(y)
    #print('gx =',gx)
    #print('gy =',gy)
    g = np.matrix([[func(gx,start_x,start_y)],[func(gy,start_x,start_y)]], dtype='float')
    #print('\n')
    #print('g =',g)
    #print('\n')
    gxx = gx.diff(x)
    gxy = gx.diff(y)
    gyx = gy.diff(x)
    gyy = gy.diff(y)
    #print('gxx =',gxx)
    #print('gxy =',gxy)
    #print('gyx =',gyx)
    #print('gyy =',gyy)
    h = np.matrix([[func(gxx,start_x,start_y),func(gxy,start_x,start_y)],[func(gyx,start_x,start_y),func(gyy,start_x,start_y)]], dtype='float')
    #print('\n')
    #print('h =',h)
    #print('\n')


    e = np.array([[1,0],[0,1]])
    s = (h + a*e)
    s = np.linalg.inv(s)
    s = s * (-g)
    #print(s)
    py = px + s
    #print('\n',py)
    py_x = float(py[0][0])
    py_y = float(py[1][0])
    fy = sp.re(func(fx,py_x,py_y))
    fx_v = sp.re(func(fx,start_x,start_y))
    #print('\n',fy)
    #print('\n',fx_v)
    if fy >= fx_v:
      a = a/b
      continue
    delta_x = py_x - start_x
    delta_y = py_y - start_y
    delta_p = sqrt(delta_x**2 + delta_y**2)/sqrt(start_x**2 + start_y**2)   

    start_x = py_x
    start_y = py_y
    a = a * b
    gl = sqrt(func(gx,start_x,start_y)**2 + func(gy,start_x,start_y)**2)
    #print('Pog1',fy - fx_v)
#    if gl <= el:
#      print('Amount of it: ',it_amount)
#      break
#    if abs(fy - fx_v) <= el and delta_p <= el:
#      print('Amount of it: ',it_amount)
#      break
    if delta_p <= el:
      #print('Amount of it: ',it_amount)
      break

  
  fx_t = (100*((y - x**2)**2)) + ((1-x)**2) + r*sp.ln(y-1.5)
  delta_f = abs(func(fx_t, start_x,start_y) - func(fx_t, old_x,old_y))
  if(delta_f <= 0.001):
    print('Final point:',start_x,start_y)
    break
  r = r * 0.1
  print('r =',r)  
  old_x = start_x
  old_y = start_y
  it_amount = 0
```

```python
def differ(fx,x1,y1,step,symbol):
  ha = step
  if symbol == 'x':
    f = (func(fx,x1+ha,y1) - func(fx,x1-ha,y1))/(2*ha)
  else:
    f = (func(fx,x1,y1+ha) - func(fx,x1,y1-ha))/(2*ha)
  return float(f)  
```

Метод Марквардта(ліміти)

```python
start_x = -1
start_y = 1
a = 1
b = 0.2
el = 0.1
step = 0.0000001 
it_amount = 0
points_x = list()
points_y = list()

while(True):
  if it_amount > 50:
    break
  px = np.matrix([[start_x],[start_y]],dtype='float')
  x = sp.Symbol('x')
  y = sp.Symbol('y')
  fx = (1 - x)**2 + 100 * ((y - x**2)**2)
  gx = fx.diff(x)
  gy = fx.diff(y)
  #print('gx =',gx)
  #print('gy =',gy)
  dx = differ(fx,start_x,start_y,step,'x')
  dy = differ(fx,start_x,start_y,step,'y')
  g = np.matrix([[dx],[dy]], dtype='float')
  #print('\n')
  #print('g =',g)
  #print('\n')
  gxx = gx.diff(x)
  gxy = gx.diff(y)
  gyx = gy.diff(x)
  gyy = gy.diff(y)
  #print('gxx =',gxx)
  #print('gxy =',gxy)
  #print('gyx =',gyx)
  #print('gyy =',gyy)
  h = np.matrix([[differ(gx,start_x,start_y,step,'x'),differ(gx,start_x,start_y,step,'y')],[differ(gy,start_x,start_y,step,'x'),differ(gy,start_x,start_y,step,'y')]], dtype='float')
  #print('\n')
  #print('h =',h)
  #print('\n')

  e = np.array([[1,0],[0,1]])
  s = (h + a*e)
  s = np.linalg.inv(s)
  s = s * (-g)
  #print(s)
  py = px + s
  print('\n',py)
  py_x = float(py[0][0])
  py_y = float(py[1][0])
  fy = func(fx,py_x,py_y)
  fx_v = func(fx,start_x,start_y)
  #print('\n',fy)
  #print('\n',fx_v)
  if fy >= fx_v:
    a = a/b
    continue
  it_amount = it_amount + 1
  points_x.append(py_x)
  points_y.append(py_y)
  delta_x = py_x - start_x
  delta_y = py_y - start_y
  delta_p = sqrt(delta_x**2 + delta_y**2)/sqrt(start_x**2 + start_y**2)   

  start_x = py_x
  start_y = py_y
  a = a * b
#  if delta_p <= el:
#    print('Amount of it: ',it_amount)
#    break 
  if abs(fy - fx_v) <= el and delta_p <= el:
    print('\n\n\nAmount of iteration: ',it_amount)
    break
plt.scatter(points_x, points_y)
plt.show()
i=1
print('#','   ','x','                     ','y')
for x,y in zip(points_x,points_y):
  print(i,'   ',x,'   ',y)
  i = i + 1
```
